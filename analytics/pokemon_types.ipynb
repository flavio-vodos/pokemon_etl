```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar os dados do SQLite
conn = sqlite3.connect('../data/pokemon_data.db')
df = pd.read_sql('SELECT * FROM pokemon', conn)
conn.close()

# Visualizar os primeiros registros
print(df.head())
```

```python
# Plotar Base Experience vs. Height
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Height', y='Base Experience', hue='Types')
plt.title('Altura vs Experiência Base dos Pokémons')
plt.xlabel('Altura')
plt.ylabel('Experiência Base')
plt.show()
```
